#### Exercice 10 : calculer la difference de moyen salaire entre homme et femme aux 50% des salaires les plus eleves

```python
salaire_homme = wage[femme==0]
salaire_femme = wage[femme==1]
```

```python
num_homme_50 = int(len(salaire_homme) * 0.5)
num_femme_50 = int(len(salaire_femme) * 0.5)
```

```python
salaire_homme_50 = salaire_homme.sort_values(ascending=False)[:num_homme_50]
salaire_femme_50 = salaire_femme.sort_values(ascending=False)[:num_femme_50]
salaire_femme_50
```

```python
salaire_homme_50.mean()  - salaire_femme_50.mean()
```
